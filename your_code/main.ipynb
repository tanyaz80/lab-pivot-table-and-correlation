#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
```

```python
data.pivot_table(index= ['Department', 'Title'], values = ['Salary'], aggfunc = ['median', 'min','max','std'])
#As we can min and max Salary for VP in every department is the same, so std =0. Analyst salary is in range from 30 to 35, 
#and the std 3.5, (stds is a measure of how dispersed the data is in relation to the mean). 
#The interesting observation that. VP in HR has only one person representing and the std is Nan.
#In case of associate in Sales, we have 2 people represnting, with the same salary, and! the std=0.
#I wonder why.
```
